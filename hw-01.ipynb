```python
#HW#1-1.2
# Решите систему уравнений (на листочке):
# 3x – 2y + 5z = 7
# 7x + 4y – 8z = 3
# 5x – 3y – 4z = –12

import numpy as np 

d = [(3,-2,5),(7,4,-8),(5,-3,-4)]
D = np.linalg.det(d)

x = [(7,-2,5),(3,4,-8),(-12,-3,-4)]
X = np.linalg.det(x)
print(X/D)
y = [(3,7,5),(7,3,-8),(5,-12,-4)]
Y = np.linalg.det(y)
print(Y/D)
z = [(3,-2,7),(7,4,3),(5,-3,-12)]
Z = np.linalg.det(z)
print(Z/D)
```

```python
#HW#1-2.1 
# 2)
# Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
# Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)
def v_length(x,y,z):
    v_len = math.sqrt(x**2 + y**2 + z**2)
    return(v_len)

v_len = v_length(20,20,0)
print(f"Длина вектора: {v_len}")
```

```python
#HW#1-2.3
# Напишите код на Python, реализующий построение графиков:
# окружности,
# эллипса,
# гиперболы.
# def plot_circle(x1,y1,)
from pylab import rcParams
%config InlineBackend.figure_format = 'svg'
rcParams['figure.figsize'] = 5, 5
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
def draw_circle(x0,y0,R):
    x=[]
    y1=[]
    y2=[]
    x=np.linspace(x0-R, x0+R,100)
    for xi in x:
        y1.append(y0 + math.sqrt(R**2 - (xi-x0)**2))
        y2.append(y0 - math.sqrt(R**2 - (xi-x0)**2))
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.axis([-5, 5.1, -5.1, 5.1])
    plt.show()
    
draw_circle(1.0,1.0,2.5)
```

```python
def draw_elipse(x0,y0,a,b):
    x=[]
    y1=[]
    y2=[]
    x=np.linspace(x0-a, x0+a,100)
    for xi in x:
        y1.append(y0 + math.sqrt(b*(1 - ((xi-x0)/a)**2)))
        y2.append(y0 - math.sqrt(b*(1 - ((xi-x0)/a)**2)))
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.axis([-4.1, 4.1, -4.1, 4.1])
    plt.show()
    
draw_elipse(-1.0,1.0,3.0,1.0)
```

```python
def draw_hyper(x0,y0,a,b):
    y1=[]
    y2=[]
    x1=np.linspace(-10, x0-a,100)
    for xi in x1:
        y1.append(y0 + b*math.sqrt(((xi-x0)/a)**2 - 1))
        y2.append(y0 - b*math.sqrt(((xi-x0)/a)**2 - 1))
    y3=[]
    y4=[]
    x2=np.linspace(x0+a, 10,100)
    for xi in x2:
        y3.append(y0 + b*math.sqrt(((xi-x0)/a)**2 - 1))
        y4.append(y0 - b*math.sqrt(((xi-x0)/a)**2 - 1))
    plt.plot(x1, y1)
    plt.plot(x1, y2)
    plt.plot(x2, y3)
    plt.plot(x2, y4)
    plt.axis([-10.1, 10.1, -10.1, 10.1])
    plt.show()
    
draw_hyper(-0.0,0.0,2.1,2.1)
```

```python
# HW#1-3.3
# 3. Задание (в программе)
# Напишите код, который будет переводить полярные координаты в декартовы.
# Напишите код, который будет рисовать график окружности в полярных координатах.
def convert_polar_2_decart(R,alpha):
    X,Y = R*cos(alpha),R*sin(alpha)
    return X,Y
    
convert_polar_2_decart(3, np.pi/4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import math

def draw_circle_polar(R):
    alpha = 0
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.axis([-8, 8, -5.5, 5.5])
    
    while alpha < 2*np.pi:
        alpha += np.pi/90
        x,y = convert_polar_2_decart(R, alpha)
        plt.scatter(x, y, 1, 'red')
    plt.show()

draw_circle_polar(3)
```
