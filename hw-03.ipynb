```python
# HW#3.3
# 3. Задание (в программе)
# Напишите код, который будет переводить полярные координаты в декартовы.
# Напишите код, который будет рисовать график окружности в полярных координатах.
def convert_polar_2_decart(R,alpha):
    X,Y = R*cos(alpha),R*sin(alpha)
    return X,Y
    
convert_polar_2_decart(3, np.pi/4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import math

def draw_circle_polar(R):
    alpha = 0
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.axis([-8, 8, -5.5, 5.5])
    
    while alpha < 2*np.pi:
        alpha += np.pi/90
        x,y = convert_polar_2_decart(R, alpha)
        plt.scatter(x, y, 1, 'red')
    plt.show()

draw_circle_polar(3)
```

```python
# HW#3.3 1)
# 4. Задание (в программе)
# Решите систему уравнений:
# y = x2 – 1
# exp(x) + x∙(1 – y) = 1

from pylab import rcParams
%config InlineBackend.figure_format = 'svg'
rcParams['figure.figsize'] = 5, 5
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

x=[]
y1=[]
y2=[]
x=np.linspace(-10, 10)
for xi in x:
    y1.append(xi**2 - 1)
    y2.append(1 + (exp(xi) - 1)/xi )
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis([-4, 6, -5.1, 50.1])
plt.show()
    

from scipy.optimize import fsolve
def equations(p):
    x,y = p
    return (y - x**2 + 1, exp(x) + x*(1-y) - 1)
```

```python
x1,y1 = fsolve(equations, [-2,0])
print(f'Корень №1: x1={x1}, y1={y1}')
```

```python
x2,y2 = fsolve(equations, [2,4])
print(f'Корень №2: x2={x2}, y2={y2}')
```

```python
x3,y3 = fsolve(equations, [4,6])
print(f'Корень №3: x3={x3}, y3={y3}')
```

```python
# HW#3.3 2)
# 4. Задание (в программе)
# Решите систему уравнений и неравенств:
# y = x2 – 1
# exp(x) + x∙(1 – y) > 1

# Определим область допустимых значений:
print(f'exp(0)={exp(0)}')
```

```python
# Выражение для нахождения y(x): 
# y(x) = 1 + (exp(x) - 1)/(x)
def y(x):
    y = 1 + (exp(x) - 1)/(x)
    return y
print(f'y(-1000000)={y(-1000000)}')
print(f'y(500)={y(500)}')
```

```python
# Выражение для нахождения левой части неравенства: 
# exp(x) + x*(1-y) - 1 > 0
def nerav_left(x,y):
    nerav_left = exp(x) + x*(1-y) - 1
    return nerav_left
print(f'nerav_left(x=-1000000, y(x)) = {nerav_left(-1000000, y(-1000000))}')
print(f'nerav_left(x=500, y(500)) = {nerav_left(500, y(500))}')
```
